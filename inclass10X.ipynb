Duplicate the regression model and plot from the note 10 notebook. 



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
mtcars = pd.read_csv('Data/mtcars.csv')
mtcars
```

```python
import scipy.stats as stats
results = stats.linregress(mtcars['disp'], mtcars['mpg'])
results
```

```python
# plot a line fit with matplotlib
b, m = results.intercept, results.slope
print(b,m)

plt.scatter('disp','mpg',data=mtcars)
plt.axline(xy1=(0, b), slope=m)
plt.xlabel('engine displacement')
plt.ylabel('mpg')
```

### Problem 10.1

Complete two additional linear regression models, one predicting mpg based on the weight of the car (wt), and another predicting mpg based on the horsepower (hp) of the car. Plot the fitted lines for these models.


```python
results2 = stats.linregress(mtcars['wt'], mtcars['mpg'])
results2
```

```python
# plot a line fit with matplotlib
b1, m1 = results2.intercept, results2.slope
print(b,m)

plt.scatter('wt','mpg',data=mtcars)
plt.axline(xy1=(0, b1), slope=m1)
plt.xlabel('weight of the car')
plt.ylabel('mpg')
```

```python
results3 = stats.linregress(mtcars['hp'], mtcars['mpg'])
results3
```

```python
# plot a line fit with matplotlib
b2, m2 = results3.intercept, results3.slope
print(b2,m2)

plt.scatter('hp','mpg',data=mtcars)
plt.axline(xy1=(0, b2), slope=m2)
plt.xlabel('horsepower (hp) of the car')
plt.ylabel('mpg')
```

### Problem 10.2

Compare the residuals from each model. Devise some plotting strategy to visually compare the three sets of residuals.

```python
mtcars['prediction'] = mtcars['disp'] * m + b
mtcars['residual'] = mtcars['mpg'] - mtcars['prediction']
mtcars
```

```python
plt.scatter('disp','mpg',data=mtcars)
plt.axline(xy1=(0, b), slope=m)
plt.vlines('disp', 'mpg', 'prediction', data=mtcars, linestyles='--', color="black")
plt.xlabel('engine displacement')
plt.ylabel('mpg')
```

```python
mtcars['prediction1'] = mtcars['wt'] * m1 + b1
mtcars['residual1'] = mtcars['mpg'] - mtcars['prediction1']
mtcars
```

```python
plt.scatter('wt','mpg',data=mtcars)
plt.axline(xy1=(0, b1), slope=m1)
plt.vlines('wt', 'mpg', 'prediction1', data=mtcars, linestyles='--', color="brown")
plt.xlabel('weight of the car')
plt.ylabel('mpg')
```

```python
mtcars['prediction2'] = mtcars['hp'] * m2 + b2
mtcars['residual2'] = mtcars['mpg'] - mtcars['prediction2']
mtcars
```

```python
plt.scatter('hp','mpg',data=mtcars)
plt.axline(xy1=(0, b2), slope=m2)
plt.vlines('hp', 'mpg', 'prediction2', data=mtcars, linestyles='--', color="red")
plt.xlabel('horsepower (hp) of the car')
plt.ylabel('mpg')
```

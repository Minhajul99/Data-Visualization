## Data and Dataset Types:
**Data Type:** The dataset is **tabular data**, as it is structured in rows and columns.

**Dataset Type:** The dataset is multivariate, as it contains multiple attributes (columns) for each record (row).

### Attribute Types:
1. **age:**

    * Type: Quantitative (ordered, sequential, interval).

    * Description: Represents the age of the individual.

2. **sex:**

    * Type: Categorical (nominal).

    * Description: Represents the gender of the individual (male or female).

3. **bmi:**

    * Type: Quantitative (ordered, sequential, ratio).

    * Description: Represents the Body Mass Index (BMI) of the individual.

4. **children:**

    * Type: Quantitative (ordered, sequential, ratio).

    * Description: Represents the number of children/dependents covered by the insurance.

5. **smoker:**

    * Type: Categorical (nominal).

    * Description: Indicates whether the individual is a smoker (yes or no).

6. **region:**

    * Type: Categorical (nominal).

    * Description: Represents the region where the individual resides (southeast, southwest, northeast, northwest).

7. **charges:**

    * Type: Quantitative (ordered, sequential, ratio).

    * Description: Represents the insurance charges (cost) for the individual.

#### 1. Factors affecting insurance charges (Correlation Analysis)
A heatmap to show the correlation between numerical variables.

```python
# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Insurance Dataset")
plt.show()
```

#### 2. Regional Health Comparison (BMI by Region)
A box plot to compare BMI distributions across different regions.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x="region", y="bmi", data=df, palette="Set1")
plt.title("BMI Distribution by Region")
plt.xlabel("Region")
plt.ylabel("BMI")
plt.show()
```

#### 3. Age vs. Insurance Charges (Effect of Age on Charges)
A scatter plot with a regression line to show the relationship between age and insurance charges.

```python
plt.figure(figsize=(8, 6))
sns.regplot(x="age", y="charges", data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title("Effect of Age on Insurance Charges")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.show()
```

#### Other plot types

`plt.plot` produces a line chart, but you can use other methods to produce other types of plt:

```python
# bar chart
# x-axis values 
x = [5, 2, 9, 4, 7] 
 
# Y-axis values 
y = [10, 5, 8, 4, 2] 
 
# Function to plot 
plt.bar(x, y) 
 
# function to show the plot 
plt.show()
```

```python
#scatter plot with added title and axis labels
# x-axis values 
x = [5, 2, 9, 4, 7] 
 
# Y-axis values 
y = [10, 5, 8, 4, 2] 
 
# Function to plot 
plt.scatter(x, y) 
 
# Adding Title
plt.title("GeeksForGeeks")
 
# Labeling the axes
plt.xlabel("Time (hr)")
plt.ylabel("Position (Km)")
 
# function to show the plot 
plt.show()
```
